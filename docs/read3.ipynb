# Levantamiento de datos (Explicación técnica)


Los datos fueron obtenidos de la plataforma TWEET BINDER mediante consultas como las siguientes:

``` 
"x" (y1 OR y2 OR z1 OR z2) -is:retweet since:fecha1 until:fecha2
```
Donde los terminos manejan la siguiente correspondencia según la explicacion anteior:

```
    - x : vivir sabroso
    - y1, y2 : negro, negra
    - z1, z2 : petro, francia
```

Así, se encasilla la población bajo los tres marcos
