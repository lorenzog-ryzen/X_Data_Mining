# Levantamiento de datos (Explicación lógica)



Los datos fueron obtenidos gracias a la plataforma de tweet binder. Esta plataforma fue manejada usando el softrware web, este software responde con reportes gracias a los queries que se realizan. Dichos queries se encargan de buscar palabras claves dentro de posts de X. Dicha plataforma presenta distintos planes, el accesido para este trabajo fué el plan inicial, el cual resultaba pertintente para realizar las consultas historicas deseadas.




## Definición de expresiones y tiempo

### Expresiones

Gracias a factores como los canales sociales ya descritos y maximizando el tamaño del marco poblacional, se definieron las siguientes palabras o frases clave:
```   
    - Vivir sabroso : siendo esta una representación del epicentro de estudio.
    - negro, negra : esta palabra resulta del marco de prejuicios que queremos analizar.
    - pertro, francia : con estas palabras entramos de lleno en el marco politico y logramos segmentar la población, manteniendo por fuera usos o contextos distintos de las expresiones anteriores.   
```


Tiempos

Gracias al marco politico tomado se encasilló la población dentro de finales de 21 a medidados de 2024
