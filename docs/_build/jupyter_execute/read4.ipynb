# Limpieza y preparación de los datos

```python
import pandas as pd
import os
from unidecode import unidecode
```

## Ingesta y estructura del conjunto de datos y Eliminación de duplicados

Los datos se importan desde archivos CSV o XLSX. Se normalizan los nombres de las columnas y se tipifican los campos clave, incluyendo id, user, created_at, text y las métricas asociadas.

```python
RAW_BASE = os.path.join("..", "data", "raw")
PROCESSED_BASE = os.path.join("..", "data", "processed")
files = os.listdir(RAW_BASE)
clean_dfs = []
for file in files:
    current_df = pd.read_excel(os.path.join(RAW_BASE, file), sheet_name="posts")
    display(len(current_df), "numero de columnas antes de limpieza")
    current_cleaned_df = current_df[["Username", "post",'UTC Date', "Likes count", "reposts"]].drop_duplicates(subset=["Username", "post"])
    display(len(current_cleaned_df), "numero de columnas despues de limpieza")
    clean_dfs.append(current_cleaned_df)
unified_df = pd.concat(clean_dfs)
unified_df = unified_df.sort_values(by='Likes count')
display("UNIFIED DF \n",unified_df)
```

## Normalización del texto

El contenido textual se normaliza aplicando Unicode NFC, convirtiendo todo el texto a minúsculas y eliminando o estandarizando las URLs.

```python
lista_posts_limpios = []
for post in unified_df["post"]:
    cleaned_post = unidecode(post.lower())
    lista_posts_limpios.append(cleaned_post)

unified_df["post"] = lista_posts_limpios
```

```python
unified_df.to_excel(os.path.join(PROCESSED_BASE, "usuarios_posts_fechas_.xlsx"), index=False )
```
